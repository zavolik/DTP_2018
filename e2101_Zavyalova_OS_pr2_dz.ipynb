```python
# или же так:
march_birth_lit_prize = df[(df['Месяц рождения'] == 'Mar') | (df['Область науки'] == 'literature')]
march_birth_lit_prize.shape[0]
# в этом случае исключено совпадение двух условий (март и литература)
```
